La variabile 'Genres' presenta un numero di valori molto alto qundi rappresentero graficamente solo i primi 30 

```python
def grafico_frequenza_assoluta(df, serie):
    plt.figure(figsize=(10,12))
    sns.countplot(data= df, y=serie, palette= 'magma', order= df[serie].value_counts().index)
    plt.title(f'Distribuzione(%) {serie}')
    plt.xlabel('Frequenza Assoluta')
    plt.ylabel('')
```

```python
def grafico_frequenza_relativa(df, serie):
    relativa = (df[serie].value_counts() / len(serie) * 100).round(2)
    
    plt.figure(figsize=(10,12))
    sns.countplot(data= df, y=serie, palette= 'magma', order=relativa.index)
    plt.title(f'Distribuzione(%) {serie}')
    plt.xlabel('Frequenza Relativa(%)')
    plt.ylabel('')
    
```

```python
def tabella_frequenze(df, serie):
    return pd.concat([df[serie].value_counts(), round(df[serie].value_counts() / len(df) * 100, 2)], axis= 1, keys= ['Assoluta' , 'Relativa(%)']).reset_index().rename(columns={'index': serie})
```

```python
grafico_frequenza_assoluta(df, 'Category')
```

```python
grafico_frequenza_relativa(df, 'Category')
```

```python
tabella_frequenze(df, 'Category')
```

```python
tabella_frequenze(df, 'Type')
```

```python
grafico_frequenza_assoluta(df, 'Content Rating')
```

```python
grafico_frequenza_relativa(df, 'Content Rating')
```

```python
tabella_frequenze(df, 'Content Rating')
```

```python
tabella_frequenze(df, 'Genres')
```

```python
df_top_30_genres = pd.concat([df['Genres'].value_counts().head(30), (df['Genres'].value_counts().head(30) / 30 * 100).round(2)], axis= 1, keys= ['Assoluta', 'Relativa']).reset_index().rename(columns= {'index': 'Genres'})
```

```python
df_top_30_genres
```

```python
plt.figure(figsize=(10,12))
sns.barplot(data= df_top_30_genres, x= 'Assoluta', y= 'Genres', palette= 'magma')
plt.title('Distribuzione Top 30 genres')
plt.xlabel('')
plt.ylabel('')
```

```python
plt.figure(figsize=(10,12))
sns.barplot(data= df_top_30_genres, x= 'Relativa', y= 'Genres', palette= 'magma')
plt.title('Distribuzione Top 30 genres')
plt.xlabel('Frequenza Relativa(%)')
plt.ylabel('')
```
